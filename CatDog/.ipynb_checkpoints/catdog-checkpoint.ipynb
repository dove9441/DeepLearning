#### 딥러닝 전 학습할 데이터들을 분류하기

```python
import os
import shutil # 파일 옮기기 쉽게 하기 위한 모듈

# 버그있어서 폴더 코드로 생성
# os.mkdir('./dataset')
# os.mkdir('./dataset/cat')
# os.mkdir('./dataset/dog')


for i in os.listdir('./dogs-vs-cats-redux-kernels-edition/train/train'):
    if 'cat' in i:
        shutil.copyfile('./dogs-vs-cats-redux-kernels-edition/train/train/'+i, './dataset/cat/'+i) # 대상 위치의 파일 -> 이동할 위치의 파일
    if 'dog' in i:
        shutil.copyfile('./dogs-vs-cats-redux-kernels-edition/train/train/'+i, './dataset/dog/'+i)
```

#### 이미지 전처리하기

```python
import tensorflow as tf

train_ds = tf.keras.preprocessing.image_dataset_from_directory('./dataset/',
                                                               image_size=(64,64),
                                                               batch_size=64, # 이미지 2만장 한번에 안 넣고 64개씩 넣기
                                                               subset='training', 
                                                               validation_split=0.2, # 윗줄이랑 같이 써서 해당 비율로 쪼개짐 validation할 거 몇개 뽑아놓기
                                                               seed=12345, # 랜덤 시드 이것도 써야 오류안남
                                                              )

val_ds = tf.keras.preprocessing.image_dataset_from_directory('./dataset/',
                                                               image_size=(64,64),
                                                               batch_size=64, # 이미지 2만장 한번에 안 넣고 64개씩 넣기
                                                               subset='validation', 
                                                               validation_split=0.2, # 윗줄이랑 같이 써서  validation할 거 몇개 뽑아놓기
                                                               seed=12345,
                                                              ) 
### 이미지라면 0~255를 0~1로 압축하면 연산 속도 및 정확도가 높아짐
def prepro(i,y):
    i = tf.cast(i/255.0, tf.float32)
    return i, y

train_ds = train_ds.map(prepro)
val_ds = val_ds.map(prepro)


# import matplotlib.pyplot as plt  
# 사진 미리보기
# for i, y in train_ds.take(1):
#     plt.imshow( i[50].numpy().astype('uint8'))
#     plt.colorbar()
#     plt.show()
#     print(i[50], y)

# train_ds와 val_ds는 학습 데이터만 다른 동일한 형식인데, (학습데이터, 정답값) 이런 튜플이 담겨져 있다.
# cat폴더와 dog폴더로 나눠놓았었는데 알아서 2 classes라고 분류를 해놓는다.
```

#### 모델 만들고 학습시키기

```python
model = tf.keras.Sequential([
    ### 모델에 넣기 전에 이미지 증강하기
    tf.keras.layers.experimental.preprocessing.RandomFlip('horizontal',input_shape=(64,64,3)), 
    # 50%확률로 뒤집어짐 input shape은 첫 번째 레이어에 넣어야 함
    tf.keras.layers.experimental.preprocessing.RandomRotation(0.1), # 돌림
    tf.keras.layers.experimental.preprocessing.RandomZoom(0.1),
    tf.keras.layers.Conv2D(32, (3,3), padding="same", activation="relu",  ), # 컬러니까 마지막이 3임
    tf.keras.layers.MaxPooling2D( (2,2) ),
    tf.keras.layers.Conv2D(64, (3,3), padding="same", activation="relu"), # 컬러니까 마지막이 3임
    tf.keras.layers.MaxPooling2D( (2,2) ),
    tf.keras.layers.Dropout(0.2), #  윗 레이어의 노드 일부 퍼센트를 제거함(20%) 왜? overfitting 막기위해
    tf.keras.layers.Conv2D(128, (3,3), padding="same", activation="relu" ), # 컬러니까 마지막이 3임
    tf.keras.layers.MaxPooling2D( (2,2) ),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dropout(0.2), #  윗 레이어의 노드 일부 퍼센트를 제거함(20%) 왜? overfitting 막기위해
    tf.keras.layers.Dense(1, activation="sigmoid"), # BCEE에서는 마지막이 sigmoid다 왠지는모름
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])

model.summary()

model.fit(train_ds, validation_data=val_ds, epochs=5)

```

#### 모델 저장 / 로드하는 방법

```python
model.save("./savedModels/model2") ## 모델 저장하기
model2 = tf.keras.models.load_model("./savedModels/model2") # 모델 불러오기
```
